```python
import pandas as pd
import json
import openpyxl
```

```python
filename = "../src/vet.js"
```

```python
ajson = {}

with open(filename, mode="r", encoding="utf-8") as f:
    atext = f.read()
    atext = atext.replace("var vet = ", "")
    ajson = json.loads(atext)
```

```python
ajson
```

```python
len(ajson["features"])
```

```python
df_lt = pd.read_json(json.dumps([ x["properties"] for x in ajson["features"] ]))
df_lt
```

```python
df_rt = pd.read_json(json.dumps([ x["geometry"]["coordinates"] for x in ajson["features"] ]))
df_rt.columns = ["Longitude", "Latitude"]
df_rt
```

```python
df = df_lt.join(df_rt)
df
```

```python
df.to_csv("output/vet_from_geojson.tsv", sep="\t", quoting=1, index=False)
```

```python
df.to_excel("output/vet_from_geojson.xlsx", index=False, engine="openpyxl")
```
