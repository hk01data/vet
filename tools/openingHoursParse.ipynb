```python
def isChinese(in_s):
    s = in_s
    if -1 != s.find("星") or \
        -1 != s.find("午") or \
        -1 != s.find("晚") or \
        -1 != s.find("至"):
        return True
    else:
        return False
```

```python
def non7Days(in_s):
    s = in_s.lower()
    if -1 != s.find("mon") or \
        -1 != s.find("sun") or \
        -1 != s.find("fri") or \
        -1 != s.find("星期"):
        return True
    else:
        return False
```

```python
def hasOff(in_s):
    s = in_s.lower()
    if -1 != s.find("off") or \
        -1 != s.find("休息"):
        return True
    else:
        return False
```

```python
def replace_char(in_s):
    s = in_s.replace("~", "-")
    s = s.replace(" ", "")
    return s
```

```python
def extract_time_range(in_s, day):
    sarr = in_s.split("-")
    
    t = []
    for item in sarr:
        hh = "-1"
        mm = "-1"
        xarr = []
        if len(item) == 4:
            hh = item[:2]
            mm = item[2:]
        else:
            xarr = item.split(":")
            is24hr = True
            """
            TODO:
            parsing on these string has issues:
            0:00AM
            0:00PM
            12:00PM
            12:00AM
            """
            if -1 != item.lower().find("pm"):
                is24hr = False

            if len(xarr) == 2:
                hh = xarr[0]
                mm = xarr[1]
                if len(hh) == 1:
                    hh = f"0{hh}"
                if len(mm) > 2:
                    mm = mm[:2]
                if not is24hr:
                    if int(hh) < 13:
                        hh = str(int(hh) + 12)
                    if hh == "24":
                        hh = "00"
            if item.lower() == "12:00pm" or item.lower() == "12:00nn" or item.lower() == "12:00noon":
                hh = "12"
                mm = "00"
            if item.lower() == "12:00am" or item.lower() == "0:00am" or item.lower() == "0:00pm":
                hh = "00"
                mm = "00"
        t.append({
              "day": day,
              "time": f"{hh}{mm}"
            })
        
    return {
        "close": t[1],
        "open": t[0]
    }

    return {
    "close": {
      "day": 5,
      "time": "2030"
    },
    "open": {
      "day": 5,
      "time": "1030"
    }
  }
```

```python
def only_time(in_s):
    s = in_s
    if not isChinese(in_s) and not non7Days(in_s):
        x = []
        for day in range(0, 7):
            print(in_s)
            x.append(extract_time_range(in_s, day))
        s = x

    return s
```

```python
"""
Hardcode
"""
def dayrange(in_s):
    s = in_s
    if type(s) != list and -1 != s.find(":") and -1 == s.find("PublicHoliday"):
        """
        TODO:
        should use regular expression to extract groups
        """
        s = s.replace("星期一至星期日(Mon-Sun)", "")
        
        x = []
        for day in range(0, 7):
            print(in_s)
            try:
                x.append(extract_time_range(s, day))
            except:
                s = in_s
        s = x
    return s
```

```python
def tttt_tt_tt(in_s):
    return f'{in_s["open"]["time"][:2]}:{in_s["open"]["time"][2:]} - {in_s["close"]["time"][:2]}:{in_s["close"]["time"][2:]}'
```

```python
new_oh = []

days = [
    f"星期一: 休息",
    f"星期二: 休息",
    f"星期三: 休息",
    f"星期四: 休息",
    f"星期五: 休息",
    f"星期六: 休息",
    f"星期日: 休息",
]

for item in oh_arr:
    s_days = days[:]
    if "close" not in item[0]:
        s_days = [
                "星期一: 24 小時營業",
                "星期二: 24 小時營業",
                "星期三: 24 小時營業",
                "星期四: 24 小時營業",
                "星期五: 24 小時營業",
                "星期六: 24 小時營業",
                "星期日: 24 小時營業",
            ]
    else:
        for d in item:
            s_days[d["open"]["day"]] = s_days[d["open"]["day"]].replace("休息", tttt_tt_tt(d))
        
    new_oh.append(s_days)
```
